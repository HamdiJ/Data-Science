```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.arange(7,58,10)
x
```

```python
x = x.reshape((-1, 1))
```

```python
x
```

```python
import matplotlib
import matplotlib.pyplot as plt
```

```python
plt.scatter(x, y)
plt.show()
```

```python
model = LinearRegression()
```

```python
model.get_params()
```

```python
model.fit(x, y)
```

```python
Coef_dÃ©t = model.score(x, y)
print('Coefficient de dÃ©termination:', Coef_dÃ©t)
```

```python
print('OrdonnÃ©e Ã  l origine:', model.intercept_)
print('pente:', model.coef_)
```

```python
y_pred = model.intercept_ + model.coef_ * x
print('prÃ©dictions:', y_pred, sep='\n')
```

```python
y_pred = model.predict(x)
print('prÃ©dictions:', y_pred, sep='\n')
```

```python
plt.scatter(x, y)
plt.plot(x, y_pred)
plt.show()
```

Retrouver une fonction polynomiale

Plus on rajoute un degrÃ© au polynÃ´me, plus ce dernier devient sensible aux donnÃ©es et sây adapte mieux.
trouvons un modÃ¨le avec un polynÃ´me de 4 degrÃ©s.


```python
x = np.array(tempsChargementPages)
y = np.array(montantAchat)

```

```python
n=4
Coff=np.polyfit(x, y, n) # ax
fp = np.poly1d(Coff)

axes = plt.axes()
axes.grid()
xp = np.linspace(0, 7, 100)
plt.scatter(x, y)
plt.plot(xp, fp(xp), c='r')
plt.show()

```
